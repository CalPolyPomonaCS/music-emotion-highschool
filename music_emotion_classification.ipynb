**Best starting dataset:**

DEAM: Database for Emotional Analysis of Music. It contains 1,802 music excerpts/full songs with valence and arousal emotion annotations.
Valence means positive/negative feeling; arousal means energy/intensity. (CVML)

**Good advanced dataset:**

PMEmo contains 794 pop-music chorus clips annotated by 457 listeners, with valence-arousal labels and physiological EDA signals. (GitHub)

**Piano-focused dataset:**

EMOPIA contains 1,087 pop piano clips with emotion labels, useful because the student is a violinist/musician and may want to analyze
